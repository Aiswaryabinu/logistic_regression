```python
# Generate a small dataset with two features
df = pd.DataFrame({'v1': [1,1,1,2,2,2,3,3,3],
                   'v2': [100,200,300,100,200,300,100,200,300]
                  })
```

```python
# add the two features together
df['v1 + v2'] = df['v1'] + df['v2']

# multiply the two features together
df['v1 x v2'] = df['v1'] * df['v2']
df
```

```python
import seaborn as sns
```

```python
df_add = df.pivot(index='v1',
                  columns='v2',
                  values='v1 + v2'
                 )
print("v1 + v2\n")
display(df_add)
print()
sns.heatmap(df_add);
```

```python

df_mult = df.pivot(index='v1',
                  columns='v2',
                  values='v1 x v2'
                 )
print('v1 x v2')
display(df_mult)
print()
sns.heatmap(df_mult);
```

Addition	Poor — may be dominated by larger variable

Multiplication	Good — captures influence of both variables

Multiplying features helps separate the feature space into a grid-like structure, useful for machine learning.

Interaction terms are powerful features that can model combined effects.
